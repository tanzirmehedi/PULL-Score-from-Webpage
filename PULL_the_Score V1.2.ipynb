```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
base_url = 'https://snyk.io/advisor/packages/python/popular'
base_url
```

```python
concatenated_urls = []
str = "https://snyk.io/advisor/python/"

def scrape_page(url):
    c=1
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    ul_element = soup.find('ul', class_='list-items')
    if ul_element:
        list_items = ul_element.find_all('li')
        for li in list_items:
            concatenated_urls.append(f"{str}{li.text}")
            print(f'Package: {c} {li.text}')
            c = c+1
    else:
        print(f'Specific <ul> element not found on the webpage: {url}')

scrape_page(base_url)
page_number = 2
while page_number < 32:
    next_page_url = f"{base_url}?page={page_number}"
    response = requests.get(next_page_url)
    if response.status_code == 200:
        scrape_page(next_page_url)
        page_number += 1
    else:
        print(f"No more pages found. Exiting...")
        break
```

```python
urls = concatenated_urls
with open('output.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Package URL', 'Package Health Value'])
    for url in urls:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            element = soup.find('div', class_='number')
            if element:
                package_health_value = element.text
                csv_writer.writerow([url, package_health_value])
                print(f'Package: {url}: {package_health_value}')
            else:
                csv_writer.writerow([url, 'Value not found'])
                print(f'Value not found on {url}')
        else:
            csv_writer.writerow([url, f'Failed to retrieve data. Status code: {response.status_code}'])
            print(f'Failed to retrieve data from {url}. Status code: {response.status_code}')

print('Data has been saved to output.csv')
```
